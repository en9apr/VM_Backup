## Why do we need Proper Orthogonal Decomposition?

* Convert high dimensional data to lower dimensional data for visualisation
* Data compression of large data sets
* Identification of instantaneous and time-varying flow structures - based on energy content
* Mode shapes or basis functions for Galerkin projections 

## What is the difference between POD and DMD?

* Dynamic mode decomposition identifies structures based on frequency content
* Proper orthogonal decomposition identifies structures based on energy content

## Which version of POD is suitable?

* For temporally resolving data - use classical method
* For spatially resolving data - Method of Snapshots

## What are the inputs?

We have the option of either:

Vector quantities, e.g. Velocity distributions or vorticity fields:

$ W^{(k)} = (u, v)_{i,j}^{(k)} $

Or:

Scalar quantities, e.g. Intensity values or snapshots:

$ s^{(k)} = (s)_{i,j}^{(k)} $

where: $(i, j)$ is the grid point index and $k$ is the snapshot index

## How is the input ordered?

The $u$ and $v$ velocities are reordered and put into a matrix:

$U = \begin{bmatrix}u^{(1)} \\u^{(2)} \\\vdots \\u^{(K)} \end{bmatrix} $

$U = \begin{bmatrix}u^{(1)}_{i=1,j=1} & \dots & u^{(1)}_{i=1,j=J} & u^{(1)}_{i=2,j=1} & \dots & u^{(1)}_{i=2,j=J} & u^{(1)}_{i=I,j=1} & \dots & u^{(1)}_{i=I,j=J} \\u^{(2)}_{i=1,j=1} & \dots & u^{(2)}_{i=1,j=J} & u^{(2)}_{i=2,j=1} & \dots & u^{(2)}_{i=2,j=J} & u^{(2)}_{i=I,j=1} & \dots & u^{(2)}_{i=I,j=J} \\\vdots & \ & \vdots & \vdots & \ & \vdots & \vdots & \ & \vdots \\u^{(K)}_{i=1,j=1} & \dots & u^{(K)}_{i=1,j=J} & u^{(K)}_{i=2,j=1} & \dots & u^{(K)}_{i=2,j=J} & u^{(K)}_{i=I,j=1} & \dots & u^{(K)}_{i=I,j=J}  \end{bmatrix}$

where: $I$ is the total number of horizontal grid points, $J$ is the total number of vertical grid points, $K$ is the total number of snapshots

## What are the outputs?

Decompose the vector or scalar field into a linear combination of:

* $M$ spatial basis functions (POD modes) $\varphi_m$

* $M$ coefficients $c_m^{(k)}$

Recompose the vector or scalar field for the $k^{th}$ snapshot:

$ W^{(k)} = \sum\limits_{m = 1}^Mc_m^{(k)}\varphi_m $

where: $M$ is the number of modes and $k$ is the snapshot number

Can also look at ensemble average field (assuming the snapshots apply to a steady-state set):

$\langle W \rangle = \dfrac 1K \sum\limits_{k = 1}^K W^{(k)} = \sum\limits_{m = 1}^M \left[\dfrac 1K \sum\limits_{k = 1}^K c_m^{(k)}\right] \varphi_m = \sum\limits_{m = 1}^M \langle c_m \rangle ^K \varphi_m$


```python
import numpy as np
x=np.linspace(0,1,15)   #array with indices i
y=np.linspace(0,2,10)   #array with indices j
t=np.linspace(0,3,5)   #array with indices k
```

```python
[X,Y]=meshgrid(x,y,indexing='ij')
```

```python
U=zeros((len(t),len(y)*len(x)))   #initialise 2D array for velocity
V=zeros((len(t),len(y)*len(x)))   #initialise 2D array for velocity
```

```python
print U
```

```python
for k in range(len(t)):                            #go from 0 to 4
    for i in range(len(x)):                        #go from 0 to 14
        for j in range(len(y)):                    #go from 0 to 9
            U[k,j+i*len(y)]=Y[i,j]+t[k]+X[i,j]
            V[k,j+i*len(y)]=Y[i,j]+t[k]+X[i,j]
```
