### 2.0 Integral Form of the Conservation of Momentum ###

**2.0.1** For a system:

$$ \vec F = {D \over D t} \int_{SYS} \vec V dM $$

**2.0.2** For a control volume (via Reynolds Transport Theorem):
    
$$ \sum \vec F_{CV} = {\partial \over \partial t} \int_{CV} \vec V \rho dV + \int_{CS} \vec V \rho \vec V \cdot \hat n dA  $$ 

### 2.1 Differential Form of the Conservation of Momentum ###

**2.1.0** For the left-hand side of the equation there are:

*2.1.0.1* Body forces - weight for an element $ \delta m$:

$$ \delta \vec F = {D \over D t} \vec V \delta m = \delta m {D \over {Dt}} \vec V = \delta m \cdot \vec a$$

$$ \delta \vec F_b = \delta m \cdot \vec g $$

*2.1.0.2* Normal force ($\delta F_n$ : normal to $\delta A$)


*2.1.0.3* Tangential force ($\delta F_1$, $\delta F_2$ : tangential to $\delta A$)

*2.1.0.4* Normal stress

$$\sigma_n = \lim_{\delta A \to 0} {\delta F_n \over \delta A}$$

*2.1.0.5* Tangential stress

$$\tau_1 = \lim_{\delta A \to 0} {\delta F_1 \over \delta A}$$

$$\tau_2 = \lim_{\delta A \to 0} {\delta F_2 \over \delta A}$$

*2.1.0.6* With reference to a coordinate system:

*2.1.0.7* Normal stress : $\sigma_{xx}$ 

*2.1.0.8* Shear stress : $\tau_{xy}$ $\tau_{xz}$

*2.1.0.9* Subscript notation: 1st subscript refers to the direction of the normal vector

*2.1.0.10* Subscript notation: 2nd subscript refers to the direction of the stress vector   

*2.1.0.11* Sign convention: Normal Stress is positive if it's in the same direction as the outward normal vector

*2.1.0.12* Sign convention: Shear Stress is positive if it's in the same direction as the coordinate system w.r.t the outward normal vector
